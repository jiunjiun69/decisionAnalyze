```python
data = pd.read_csv('data.csv')
print(data.shape)
print(data.info())
data
```

```python
sum = 0
for i in range(100,1001,100):
    for j in range(i+100,1001,100):
        for k in range(j+100,1001,100):
            for w in range(k+100,1001,100):
                sum+=1
                print(i,j,k,w)
print('總共:' + str(sum) + '種組合')
```

```python
time_max = []
time_8 = []
time_4 = []
time_2 = []
```

```python
sum = 0
for i in range(100,1201,100):
    for j in range(i+100,1201,100):
        for k in range(j+100,1201,100):
            for w in range(k+100,1201,100):
                time_max.append(i)
                time_8.append(j)
                time_4.append(k)
                time_2.append(w)
                sum+=1
                print(i,j,k,w)
print('總共:' + str(sum) + '種組合')
```

```python
combinations = pd.DataFrame(columns=['一般維修','8個小時','4個小時','2個小時'])

combinations['一般維修'] = time_max
combinations['8個小時'] = time_8
combinations['4個小時'] = time_4
combinations['2個小時'] = time_2

combinations
```

```python
# 平均接受度
accept = []

# 平均獲利
profit = []

for i in range(len(combinations)):
    accept.append(\
    float(data[data['價格']==combinations['一般維修'][i]]['一般維修'])*0.25 +\
    float(data[data['價格']==combinations['8個小時'][i]]['8個小時'])*0.25 +\
    float(data[data['價格']==combinations['4個小時'][i]]['4個小時'])*0.25 +\
    float(data[data['價格']==combinations['2個小時'][i]]['2個小時'])*0.25)
    
    profit.append(\
    float(data[data['價格']==combinations['一般維修'][i]]['一般維修'])*0.25*float(combinations['一般維修'][i]) +\
    float(data[data['價格']==combinations['8個小時'][i]]['8個小時'])*0.25*float(combinations['8個小時'][i]) +\
    float(data[data['價格']==combinations['4個小時'][i]]['4個小時'])*0.25*float(combinations['4個小時'][i]) +\
    float(data[data['價格']==combinations['2個小時'][i]]['2個小時'])*0.25*float(combinations['2個小時'][i]))
    
combinations['平均接受度'] = accept
combinations['平均獲利'] = profit
combinations
```
